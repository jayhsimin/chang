```python
# 相除+排序
cpath=r'C:\Users\vghuser\Desktop\xxx\all_s'
ccc=r'C:\Users\vghuser\Desktop\area\1&2'
x12=os.path.join(cpath,'divide')
if not os.path.isdir(x12):
    os.mkdir(x12)
file1 = glob(f'{cpath}/x12_*.csv') 
file2 = glob(f'{ccc}/x1_*.csv')    

for i in range(6):
    e=file1[i]
    f=file2[i]

    # xai
    basenames = os.path.basename(e)
    bk = os.path.splitext(basenames)[0]
    aa=bk.split('_')[0]
    bb=bk.split('_')[1]

    #bin
    basenamess = os.path.basename(f)
    bks = os.path.splitext(basenamess)[0]
    AA=bks.split('_')[0]
    BB=bks.split('_')[1]

    if bb==BB:
        if bb=='4r' or bb=='4s':
            df = pd.merge(pd.read_csv(e),pd.read_csv(f) , on='id1')
            for i in range(1,5):
                df[f'XW{i}'] = df[f'XW{i}'] / (df[f'area{i}']/1000) 
            pd.DataFrame(df,columns=['id1','XW1','XW2','XW3','XW4'])
            rank_list = []
            for ID in df['id1']:
                # print(f'‧ now searching : {ID}')
                temp_Q_list = []
                temp_num_list = []
                temp_str = ''
                for Q in ['XW1','XW2','XW3','XW4']:#4
                    temp_Q_list.append(Q)
                    temp_num_list.append(df[df['id1']==ID][Q].reset_index(drop=True)[0])
                    temp_df = pd.DataFrame({'Q': temp_Q_list, 'num': temp_num_list})
                    temp_df = temp_df.sort_values(by = 'num', ascending = False)        
                for string in temp_df['Q']:
                    temp_str += string
                rank_list.append(temp_str)

            df['xai_rank'] = rank_list                    
            df.to_csv(f'{x12}/{bk}.csv',columns=['id1','XW1','XW2','XW3','XW4','xai_rank'],index=False)
            
        if bb=='8r' or bb=='8s':
            df = pd.merge(pd.read_csv(e),pd.read_csv(f) , on='id1')
            for i in range(1,9):
                df[f'XW{i}'] = df[f'XW{i}'] / (df[f'area{i}']/1000)
           

            pd.DataFrame(df,columns=['id1','XW1','XW2','XW3','XW4','XW5','XW6','XW7','XW8'])#8
            rank_list = []
            for ID in df['id1']:
                # print(f'‧ now searching : {ID}')
                temp_Q_list = []
                temp_num_list = []
                temp_str = ''
                lis=['XW1','XW2','XW3','XW4','XW5','XW6','XW7','XW8']
                for Q in lis:
                    temp_Q_list.append(Q)
                    temp_num_list.append(df[df['id1']==ID][Q].reset_index(drop=True)[0])
                    temp_df = pd.DataFrame({'Q': temp_Q_list, 'num': temp_num_list})
                    temp_df = temp_df.sort_values(by = 'num', ascending = False)        
                for string in temp_df['Q']:
                    temp_str += string
                rank_list.append(temp_str)

            df['xai_rank'] = rank_list 
            df.to_csv(f'{x12}/{bk}.csv',columns=['id1','XW1','XW2','XW3','XW4','XW5','XW6','XW7','XW8','xai_rank'],index=False)            

        if bb=='12r' or bb=='12s':
            df = pd.merge(pd.read_csv(e),pd.read_csv(f) , on='id1')
            for i in range(1,13):
                df[f'XW{i}'] = df[f'XW{i}'] / (df[f'area{i}']/1000)   
            pd.DataFrame(df,columns=['id1','XW1','XW2','XW3','XW4','XW5','XW6','XW7','XW8','XW9','XW10','XW11','XW12'])#12
            rank_list = []
            for ID in df['id1']:
                # print(f'‧ now searching : {ID}')
                temp_Q_list = []
                temp_num_list = []
                temp_str = ''
                lis=['XW1','XW2','XW3','XW4','XW5','XW6','XW7','XW8','XW9','XW10','XW11','XW12']
                for Q in lis:
                    temp_Q_list.append(Q)
                    temp_num_list.append(df[df['id1']==ID][Q].reset_index(drop=True)[0])
                    temp_df = pd.DataFrame({'Q': temp_Q_list, 'num': temp_num_list})
                    temp_df = temp_df.sort_values(by = 'num', ascending = False)        
                for string in temp_df['Q']:
                    temp_str += string
                rank_list.append(temp_str)
            df['xai_rank'] = rank_list                           
            df.to_csv(f'{x12}/{bk}.csv',columns=['id1','XW1','XW2','XW3','XW4','XW5','XW6','XW7','XW8','XW9','XW10','XW11','XW12','xai_rank'],index=False)            
```
