```python
import os
import pandas as pd
from fuzzywuzzy import fuzz
import scipy.stats
import numpy as np
from scipy.spatial.distance import hamming
import Levenshtein as LV
import Levenshtein
import os.path
import shutil
import re
import statistics
from glob import glob
# Dice coefficient distance
def bigram_sequence(text_lst):
    result = [a for ls in text_lst for a in zip(ls.split(",")[:-1], ls.split(",")[1:])]
    return result
# k-mer distance
def jaccard_similarity(a, b):
    a = set(a)
    b = set(b)

    intersection = len(a.intersection(b))
    union = len(a.union(b))

    return intersection / union
def build_kmers(sequence, ksize):
    kmers = []
    n_kmers = len(sequence) - ksize + 1
    for i in range(n_kmers):
        kmer = sequence[i:i + ksize]
        kmers.append(kmer)
    return kmers
def dice(c, d):
        c=set(c)
        d=set(d)
        return (2*len(c.intersection(d))) / (len(c) + len(d))
cpath=r'C:\Users\vghuser\Desktop\xxx\all_s\divide\xai-thi'
fin=os.path.join(cpath,'final')
if not os.path.isdir(fin):
    os.mkdir(fin)
file1 = glob(f'{cpath}/*.csv')
for file in file1:
    CK=[]
    DK=[]
    ss=[]
    tt=[]
    L_ratio=[]
    L_distance=[]
    H_distance=[]
    W_Distance=[]
    W_Distance_Chang=[]
    A_ratio=[]
    A_partial_ratio=[]
    D_distance=[]
    K1_distance=[]
    K2_distance=[]
    K3_distance=[]
    K4_distance=[]
    K5_distance=[]
    K6_distance=[]
    K7_distance=[]
    K8_distance=[]
    K9_distance=[]
    df=pd.read_csv(file)

    for i in range(df.shape[0]):
        # print(i)

        CC=re.findall(r"\d+",df['xai_rank'][i])
        DD=re.findall(r'\d+',df['t_rank'][i])
        CK.append(CC)
        DK.append(DD)
        basenamess = os.path.basename(file)
        bks = os.path.splitext(basenamess)[0]



    df['x']=CK
    df['fc']=DK

    for s in CK:
        results = list(map(int, s))
        ss.append(results)
    for t in DK:
        results = list(map(int, t))
        tt.append(results)
    df['xai']=ss
    df['xai-thi']=tt
    # 全部公式的變數


    for n in range(df.shape[0]):
        #給DICE用
        GG=tuple(ss[n])
        GF=tuple(tt[n])
        D1=[str(GG)]
        D2=[str(GF)]

        # levenshteinDistance(分數越小越好)*
        ratio = Levenshtein.ratio(GG,GF)#距離比 =（編輯距離）/（對齊長度）
        dist = Levenshtein.distance(GG,GF)#編輯次數
        L_ratio.append(ratio)
        L_distance.append(dist)


        # hamming_distance*
        hamming_distance = hamming(GG, GF)*len(GG)
        H_distance.append(hamming_distance)

        #Wasserstein Distance*
        P = np.array(GG)
        Q = np.array(GF)

        dists=[i for i in range(len(P))]
        Wasserstein_Distance=scipy.stats.wasserstein_distance(dists,dists,P,Q)
        W_Distance.append(Wasserstein_Distance)

        #張式Wasserstein Distance *
        c=tuple(abs(GG[i]-GF[i]) for i in range (0,len(GG)))
        l=0
        if range(len(c))==range(0, 4):
            for discrepancy in c:
                if discrepancy in range(0,1):
                    l=l+discrepancy*1
                if discrepancy in range(1,3):
                    l=l+discrepancy*1
                if discrepancy in range(3,5):
                    l=l+discrepancy*2
            W_Distance_Chang.append(l)
        if range(len(c))==range(0, 8):
            for discrepancy in c:
                if discrepancy in range(0,3):
                    l=l+discrepancy*2
                if discrepancy in range(3,5):
                    l=l+discrepancy*2
                if discrepancy in range(5,9):
                    l=l+discrepancy*3
            W_Distance_Chang.append(l)
        if range(len(c))==range(0, 12):
            for discrepancy in c:
                # print(discrepancy)
                if discrepancy in range(0,4):
                    l=l+discrepancy*3
                if discrepancy in range(4,8):
                    l=l+discrepancy*3
                if discrepancy in range(8,12):
                    l=l+discrepancy*5
            W_Distance_Chang.append(l)

        # Approximate string distance
        Approximate_string_distance = fuzz.ratio(GG,GF)
        partial_ratio = fuzz.partial_ratio(GG,GF)
        A_ratio.append(Approximate_string_distance)
        A_partial_ratio.append(partial_ratio)

        # k-mer distance*

        # for K in range(1,5):
        #     kmers1 = build_kmers(GG, K)
        #     kmers2 = build_kmers(GF, K)
    #         if K==1:
    #             kmer1=jaccard_similarity(kmers1, kmers2)
    #             K1_distance.append(kmer1)

    #         if K==2:
    #             kmer2=jaccard_similarity(kmers1, kmers2)
    #             K2_distance.append(kmer2)

    #         if K==3:
    #             kmer3=jaccard_similarity(kmers1, kmers2)
    #             K3_distance.append(kmer3)

    #         if K==4:
    #             kmer4=jaccard_similarity(kmers1, kmers2)
    #             K4_distance.append(kmer4)

            # if K==5:
    #             kmer5=jaccard_similarity(kmers1, kmers2)
    #             K5_distance.append(kmer5)

    #         if K==6:
    #             kmer6=jaccard_similarity(kmers1, kmers2)
    #             K6_distance.append(kmer6)

    #         if K==7:
    #             kmer7=jaccard_similarity(kmers1, kmers2)
    #             K7_distance.append(kmer7)

    #         if K==8:
    #             kmer8=jaccard_similarity(kmers1, kmers2)
    #             K8_distance.append(kmer8)

    #         if K==9:
    #             kmer9=jaccard_similarity(kmers1, kmers2)
    #             K9_distance.append(kmer9)

        # Dice coefficient distance
        c=bigram_sequence(D1)
        d=bigram_sequence(D2)
        Dice=dice(c, d)
        D_distance.append(Dice)


    df['L_ratio']=L_ratio
    df['L_distance']=L_distance
    df['H_distance']=H_distance
    df['W_Distance']=W_Distance
    df['W_Distance_Chang']=W_Distance_Chang
    df['A_ratio']=A_ratio
    df['A_partial_ratio']=A_partial_ratio
    df['D_distance']=D_distance
    # df['K1_distance']=K1_distance
    # df['K2_distance']=K2_distance
    # df['K3_distance']=K3_distance
    # df['K4_distance']=K4_distance
    # df['K5_distance']=K5_distance
    # df['K6_distance']=K6_distance
    # df['K7_distance']=K7_distance
    # df['K8_distance']=K8_distance
    # df['K9_distance']=K9_distance
    # df
    aa=df.drop(columns=['xai_rank','t_rank','x','fc'])
    # aa
    # des=aa.describe()
    # des=pd.DataFrame(des)
    # median=des.median(axis = 0)
    # des['median']=statistics.median(eval(des.iloc[4:]))
    # print(median)
    # Python (Pandas套件，簡稱pd)
    # des=aa.describe()
    # des=pd.DataFrame(des)
    # medi=[]
    # med=aa.median()#算中位數
    # medi.append(med)#中位數存入陣列
    # median=pd.DataFrame(medi,index=['median'])
    # analytics=pd.concat([des,median])

    # analytics.to_csv(r'C:\Users\vghuser\Desktop\bbbb.csv',index=True)
    # break



    aa.to_csv(f'{fin}\{bks}_a.csv',index=False)
    # print(f'{fin}\{bks}_a.csv')
```
