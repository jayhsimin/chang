```python
import numpy as np
import os
import pandas as pd
from glob import glob
```

```python
pp=r'C:\Users\vghuser\Desktop\FAG_\back\all_s'#後
tt=r'C:\Users\vghuser\Desktop\FAG_\front\all_s'#前
files = glob(f'{pp}\*.csv')    
file1s = glob(f'{tt}\*.csv') 

cpath=os.path.join(r'C:\Users\vghuser\Desktop\FAG_','all_c')
if not os.path.isdir(cpath):
    os.mkdir(cpath)

for i in range(len(files)):
    c=files[i]
    d=file1s[i]

    #後
    basenames = os.path.basename(c)
    bk = os.path.splitext(basenames)[0]
    aa=bk.split('_')[0]
    bb=bk.split('_')[1]
    #前
    basenamess = os.path.basename(d)
    bks = os.path.splitext(basenamess)[0]
    AA=bks.split('_')[0]
    BB=bks.split('_')[1]
    # print(bk,bks)    
    
    if bk == bks:  
        if aa=='4r' or aa=='4s':
            df = pd.merge(pd.read_csv(c),pd.read_csv(d) , on='id1')
            for n in range(1,5):
                df[f'W{n}'] = abs(df[f'TW{n}_x'] - df[f'TW{n}_y'])
            df.to_csv(f'{cpath}/{bk}_c.csv',columns=['id1','W1','W2','W3','W4'],index=False)
        
        # print(f'{cpath}\{aa}_{bb}_c.csv')
    if bk == bks:
        if aa=='8r' or aa=='8s':
            df = pd.merge(pd.read_csv(c),pd.read_csv(d) , on='id1')
            for n in range(1,9):
                df[f'W{n}'] = abs(df[f'TW{n}_x'] - df[f'TW{n}_y'])
            df.to_csv(f'{cpath}/{bk}_c.csv',columns=['id1','W1','W2','W3','W4','W5','W6','W7','W8'],index=False)    
    if bk == bks:
        if aa=='12r' or aa=='12s':
            df = pd.merge(pd.read_csv(c),pd.read_csv(d) , on='id1')
            for n in range(1,13):
                df[f'W{n}'] = abs(df[f'TW{n}_x'] - df[f'TW{n}_y'])            
            df.to_csv(f'{cpath}/{bk}_c.csv',columns=['id1','W1','W2','W3','W4','W5','W6','W7','W8','W9','W10','W11','W12'],index=False)
            
files = glob(f'{cpath}/*.csv')  
for file in files:
    df = pd.read_csv(file)
    basename = os.path.basename(file)
    filee = os.path.splitext(basename)[0]
    print(basename)
    newname=filee+'_s'
    # AA=filee.split('_')[0]
    # BB=filee.split('_')[1]
    # CC=filee.split('_')[2]

    # print(BB)
    if '4r'  in filee or '4s' in filee:
        # print(file)

        pd.DataFrame(df,columns=['id1','W1','W2','W3','W4'])#4
        rank_list = []
        for ID in df['id1']:
            # print(f'‧ now searching : {ID}')
            temp_Q_list = []
            temp_num_list = []
            temp_str = ''
            lis=['W1','W2','W3','W4']
            for Q in lis:
                temp_Q_list.append(Q)
                temp_num_list.append(df[df['id1']==ID][Q].reset_index(drop=True)[0])
                temp_df = pd.DataFrame({'Q': temp_Q_list, 'num': temp_num_list})
                temp_df = temp_df.sort_values(by = 'num', ascending = False)        
            for string in temp_df['Q']:
                temp_str += string
            rank_list.append(temp_str)

        df['t_rank'] = rank_list


        df.to_csv(f'{cpath}\{newname}.csv',index=False) 
    elif '8r'  in filee or '8s' in filee:
        # print(file)
        pd.DataFrame(df,columns=['id1','W1','W2','W3','W4','W5','W6','W7','W8'])#8
        rank_list = []
        for ID in df['id1']:
            # print(f'‧ now searching : {ID}')
            temp_Q_list = []
            temp_num_list = []
            temp_str = ''
            lis=['W1','W2','W3','W4','W5','W6','W7','W8']
            for Q in lis:
                temp_Q_list.append(Q)
                temp_num_list.append(df[df['id1']==ID][Q].reset_index(drop=True)[0])
                temp_df = pd.DataFrame({'Q': temp_Q_list, 'num': temp_num_list})
                temp_df = temp_df.sort_values(by = 'num', ascending = False)        
            for string in temp_df['Q']:
                temp_str += string
            rank_list.append(temp_str)

        df['t_rank'] = rank_list

        df.to_csv(f'{cpath}\{newname}.csv',index=False) 
    elif '12r'  in filee or '12s' in filee:
        pd.DataFrame(df,columns=['id1','W1','W2','W3','W4','W5','W6','W7','W8','W9','W10','W11','W12'])#12
        rank_list = []
        for ID in df['id1']:
            # print(f'‧ now searching : {ID}')
            temp_Q_list = []
            temp_num_list = []
            temp_str = ''
            lis=['W1','W2','W3','W4','W5','W6','W7','W8','W9','W10','W11','W12']
            for Q in lis:
                temp_Q_list.append(Q)
                temp_num_list.append(df[df['id1']==ID][Q].reset_index(drop=True)[0])
                temp_df = pd.DataFrame({'Q': temp_Q_list, 'num': temp_num_list})
                temp_df = temp_df.sort_values(by = 'num', ascending = False)        
            for string in temp_df['Q']:
                temp_str += string
            rank_list.append(temp_str)

        df['t_rank'] = rank_list

        df.to_csv(f'{cpath}\{newname}.csv',index=False) 
```
